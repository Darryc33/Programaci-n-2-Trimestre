Farem un programa que ens representi la recta que li indiquem:

De la recta        y = ax + b        el programa em demana el pendent (a) i l'ordenada a l'origen (b)  i un interval [xmín, xmàx], i em representa la recta en aquell interval 

Per millorar el programa, poseu títols als eixos, diferents colors.

```python
import matplotlib.pyplot as plt
import numpy as np
a=float(input('Introduce el valor de la pendiente(a): ' ))
b=float(input('Introduce el valor de la ordenada de origen(b): '))

xmin=b/a
xmax=-2*b/a
x=np.linspace(xmin,xmax,50)
y=(a*x)+b

plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(color="grey", linestyle='-')
plt.xlim(min(xmin,xmax), max(xmin,xmax))
plt.xlabel("x eje",fontsize=20)
plt.ylabel("y eje",fontsize=20)
tit="y="+str(a)+"x + "+str(b)+"\n"

plt.title(tit,fontsize=24)
plt.plot(x,y)
plt.show()
```
