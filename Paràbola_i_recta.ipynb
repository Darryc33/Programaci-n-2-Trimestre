```python
import matplotlib.pyplot as plt
import numpy as np

a=float(input('Introduce el valor de la pendiente(a): ' ))
b=float(input('Introduce el valor de la ordenada de origen(b): '))
c=float(input('Introduce el valor de c: '))
if a!=0:
  xmin=b/a  #Calcul de la x minima
  xmax=-b*2/a   #Calcul de la x màxima
  x=np.linspace(xmin,xmax,50)  #Calcul de l'eix x (per on es formara la parabola)
  y=a*(x)**2+b*x+c  #Calcul de l'eix y

else:
  xmin=b/c
  xmax=-2*b/c
  x=np.linspace(xmin,xmax,50)
  y=(c*x)+b

xv=-b/(2*a)
yv=a*(xv)**2+b*xv+c

plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(color="black", linestyle='-')
plt.xlim(xmin, xmax)
plt.xlabel('$\sqrt{\dfrac{x}{3}}$', fontsize=24, color='grey')
plt.ylabel("$",fontsize=24,color='grey')
plt.title("Gràfica",fontsize=24)
plt.plot(x,y)
plt.plot(xv,yv,'o',color='red')
plt.show()

```
